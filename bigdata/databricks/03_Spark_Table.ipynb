##### Cached the table in-memory for faster fetching/quering on that table
##### lazy: 
1. lazy will execute the cache statement very fast as it will not cache the data immediately, 
2. it will mark the table to be cached. 
3. Instead of, at the 1st time of use of the table, it will cache the whole table data into memory.

```
%sql
drop view if exists fire_service_calls_tbl_cache;
```

```
%sql
cache lazy table fire_service_calls_tbl_cache as
select * from demo_db.fire_service_calls_tbl
```

##### Basic SQL queries on the spark table

```
%sql
--first time its took 3 mints to execute, as its loaded the data into memory on 1st use of the table.
--from 2nd time onwards, it will take milliseconds time to execute.
select count(*) from demo_db.fire_service_calls_tbl;
```

##### Q1. How many distinct types of calls were made to the Fire Department?

```
%sql
select count(distinct CallType) as distinct_call_type_count 
from demo_db.fire_service_calls_tbl
where CallType is not null
```

##### Q2. What are the distinct types of calls made to the fire department?

```
%sql
select distinct CallType as distinct_call_type 
from demo_db.fire_service_calls_tbl
where CallType is not null
```

##### Q3. Find out all responses for delayed times greater than 5 mints?

```
%sql
select CallNumber, Delay
from demo_db.fire_service_calls_tbl
where Delay > 5
```

##### Q4. What were the most common call types?

```
%sql
select CallType, count(*) as count
from demo_db.fire_service_calls_tbl
where CallType is not null
group by CallType
order by count desc
```

##### Q5. What zip codes accounted for the most common calls?

```
%sql
select CallType, zipcode, count(*) as count
from demo_db.fire_service_calls_tbl
where CallType is not null
group by CallType, zipcode
order by count desc
```

##### Q6. What San Francisco neighborhoods are in zip codes 94102 and 94103

```
%sql
select zipcode, Neighborhood
from demo_db.fire_service_calls_tbl
where zipcode == 94102 or zipcode == 94103
```

##### Q7. What was the sum of all call alarms, average, min and max of the call response times?

```
%sql
select Sum(NumAlarms) as sum_call_alarms, 
avg(delay) as avg_delay, 
min(delay) as min_delay, 
max(delay) as max_delay
from demo_db.fire_service_calls_tbl
```

##### Q8. How many distinct years of data are in csv files?

```
%sql
select distinct year(to_date(CallDate, "MM/dd/yyyy")) as distinct_year
from demo_db.fire_service_calls_tbl
```

##### Q9. What week of the year in 2018 had the most fire calls?

```
%sql
select weekofyear(to_date(CallDate, "MM/dd/yyyy")) as week_year, count(*) as count 
from demo_db.fire_service_calls_tbl
where year(to_date(CallDate, "MM/dd/yyyy")) = 2018
group by week_year
order by count desc
```

##### Q10. What neighborhoods in San Francisco had the worst response time in 2018?

```
%sql
select Neighborhood, delay
from demo_db.fire_service_calls_tbl
where year(to_date(CallDate, "MM/dd/yyyy")) = 2018
order by delay desc
```
